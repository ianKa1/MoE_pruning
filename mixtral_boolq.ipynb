```python
from datasets import load_dataset
import evaluate

dataset = load_dataset("google/boolq")
metric = evaluate.load("accuracy")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_id = "kaaiiii/Mixtral_prune_1_experts" 
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto")

def predict(question, passage):
    prompt = f"Passage: {passage}\nQuestion: {question}\nAnswer yes or no."
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    out = model.generate(**inputs, max_new_tokens=5)
    ans = tokenizer.decode(out[0], skip_special_tokens=True).lower()
    return "yes" if "yes" in ans[:10] else "no"
```

```python
save_dir = "/models/mixtral_prune_1"

model.save_pretrained(save_dir)
tokenizer.save_pretrained(save_dir)
```

```python
preds = []
labels = []

for item in dataset["validation"][:1000]:  # 先测1000条
    pred = predict(item["question"], item["passage"])
    preds.append(pred == "yes")
    labels.append(item["answer"])

accuracy = metric.compute(predictions=preds, references=labels)
print(accuracy)
```
